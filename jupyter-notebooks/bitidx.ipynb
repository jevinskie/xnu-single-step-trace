```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
TBits = 16

def word_idx(sb):
    return sb // TBits

def even_inner_bit_idx(sb):
    return sb % (2 * TBits)

def odd_inner_bit_idx(sb):
    return (sb + TBits) % (2 * TBits)

def inner_bit_idx(sb):
    if word_idx(sb) % 2 == 0:
        return even_inner_bit_idx(sb)
    else:
        return odd_inner_bit_idx(sb)
```

```python
x = range(64)
y = [inner_bit_idx(i) for i in x]
```

```python
plt.plot(x, y)
```

```python
def inner_bit_idx2(sb):
    return sb - ((sb // TBits) * TBits)
```

```python
y2 = [inner_bit_idx2(i) for i in x]
```

```python
plt.plot(x, y2)
```
